```python
# Python Program to swap two variables


str1 = ("Thelma Morsindi = 16")
str2 = ("Nkechi Ikeyi = 17")

# To take inputs from the user
#str1 = input('Enter value of str1: ')
#str2 = input('Enter value of str2: ')

# displaying string str1
print(str1)
# displaying string str2
print(str2)

# tuple of mixed type elements
t1 = ("Thelma Morsindi", 16)
t2 = ("Nkechi Ikeyi", 17)

# create a temporary value and swap the values
temp = t1
t1 = t2
t2 = temp

print("The age of Thelma Morsindi after swapping=",t1[1])
print("The age of Nkechi Ikeyi after swapping=",t2[1])  
```
